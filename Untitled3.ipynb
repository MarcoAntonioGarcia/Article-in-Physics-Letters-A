```python
def theta_function(m,theta,tt): 
    alpha=0
    beta=0
    alpha2=0
    beta2=0
    gamma_inc_inf=scipy.special.gamma(3/2)*scipy.special.gammainc(3/2, (tt**2)/(4*theta));
    g00=(1-4*m*gamma_inc_inf/(np.sqrt(np.pi)*tt))
    for k in range(len(tt)):  
        if g00[k-1]>0 and g00[k]<0:
            alpha=g00[k]
            alpha2=k
        if g00[k-1]<0 and g00[k]>0:
            beta=g00[k]
            beta2=k    
    return (g00,alpha, alpha2,beta, beta2)
```

```python
tt=np.linspace(0.001, 7, 10000)
theta1= 0.0528   #Horizonte de eventos intero en r_s/2 donde r_s=1
theta2= 0.04     #Dos horizontes
theta3= 0.03     #Dos horizontes
theta4= 0.02     #Dos horizontes
theta_ext=0.06895250746595809      #Extremal BH
theta_NH=0.1                       #No horizontes BH
m=0.5
```

```python
g00_1=theta_function(m,theta1,tt)
g00_2=theta_function(m,theta2,tt)
g00_3=theta_function(m,theta3,tt)
g00_4=theta_function(m,theta4,tt)
g00_Extremal=theta_function(m,theta_ext,tt)
g00_NH=theta_function(m,theta_NH,tt)
```

```python
def geodesicas_polinomio(n,ti,tf,z0,y0,Or,mymodel_ex): 
    h=(tf-ti)/n  
    time1=list(range(0,n+1));
    y=list(range(0,n+1)) 
    z=list(range(0,n+1))
    r=list(range(0,n+1))
    y[0]=y0
    z[0]=z0 
    r[0]=1/z0
    time1[0]=ti
    
    for i in range(0,len(time1)-1):
        time1[i+1]=time1[i]+h
        l1=0 
        l2=0
        l3=0
        l4=0
        k1=h*y[i] 
        for j1 in range (Or+1):
            l1=l1+h*(-mymodel_ex[j1])*(1-j1/2)*z[i]**(1-j1) 
        
        k2=h*(y[i]+l1/2)
        for j2 in range (Or+1):
            l2=l2+h*(-mymodel_ex[j2])*(1-j2/2)*(z[i]+k1/2)**(1-j2)
        
        k3=h*(y[i]+l2/2)
        for j3 in range (Or+1):     
            l3=l3+h*(-mymodel_ex[j3])*(1-j3/2)*(z[i]+k2/2)**(1-j3)
        
        k4=h*(y[i]+l3)
        for j4 in range (Or+1):  
            l4=l4+h*(-mymodel_ex[j4])*(1-j4/2)*(z[i]+k3/2)**(1-j4)
        
            
        z[i+1]=(z[i]+(1/6)*(k1+2*k2+2*k3+k4))
        y[i+1]=y[i]+(1/6)*(l1+2*l2+2*l3+l4)
        r[i+1]=1/z[i]        
        
    x_graf=r*np.cos(time1)
    y_graf=r*np.sin(time1)

    return x_graf,y_graf,time1
```
