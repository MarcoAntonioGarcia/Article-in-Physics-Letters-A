```python
def geodesicas_polinomio(n,ti,tf,z0,y0,Or,rh,mymodel_ex): 
    h=(tf-ti)/n  
    time1=list(range(0,n+1));
    y=list(range(0,n+1)) 
    z=list(range(0,n+1))
    r=list(range(0,n+1))
    y[0]=y0
    z[0]=z0 
    r[0]=1/z0
    time1[0]=ti
    
    for i in range(0,len(time1)-1):
        time1[i+1]=time1[i]+h
        l1=0 
        l2=0
        l3=0
        l4=0
        k1=h*y[i] 
        for j1 in range (Or+1):
            l1=l1+h*(-mymodel_ex[j1])*(1-j1/2)*z[i]**(1-j1) 
        
        k2=h*(y[i]+l1/2)
        for j2 in range (Or+1):
            l2=l2+h*(-mymodel_ex[j2])*(1-j2/2)*(z[i]+k1/2)**(1-j2)
        
        k3=h*(y[i]+l2/2)
        for j3 in range (Or+1):     
            l3=l3+h*(-mymodel_ex[j3])*(1-j3/2)*(z[i]+k2/2)**(1-j3)
        
        k4=h*(y[i]+l3)
        for j4 in range (Or+1):  
            l4=l4+h*(-mymodel_ex[j4])*(1-j4/2)*(z[i]+k3/2)**(1-j4)
        
            
        z[i+1]=(z[i]+(1/6)*(k1+2*k2+2*k3+k4))
        y[i+1]=y[i]+(1/6)*(l1+2*l2+2*l3+l4)
        r[i+1]=1/z[i]        
        
    x_graf=r*np.cos(time1)
    y_graf=r*np.sin(time1)

    return x_graf,y_graf,time1
```

```python
def den_funcion(dat,malla_x,malla_y,L):
    x=dat[:,0]
    y=dat[:,1]
    z=dat[:,2]
    z_max=np.max(z)
    den=np.zeros([malla_x,malla_y])
    z_new=np.zeros([malla_x,malla_y])
    z_final=np.zeros([malla_x,malla_y])
    x=x*malla_x/L
    y=y*malla_x/L
    for i in range(len(x)):   
        if z_max!=1:
            z[i]=z[i]/z_max
        a=int(int(x[i])+malla_x/2-1)
        b=-int(int(y[i])+malla_y/2-1)
        den[a,b]=den[a,b]+1
        z_new[a,b]=z_new[a,b]+z[i]
        z_final[b,a]=z_new[a,b]/den[a,b]
    return z_final
```

```python
def geodesicas(d,l,rs,n,ti,tf,z0,y0,q): 
    h=(tf-ti)/n  
    time1=list(range(0,n+1));
    y=list(range(0,n+1)) 
    z=list(range(0,n+1))
    r=list(range(0,n+1))
    y[0]=y0
    z[0]=z0 
    r[0]=1/z0
    time1[0]=ti
    m=rs/2
    rh=rs/2+np.sqrt(rs**2-4*q**2)/2
    for i in range(0,len(time1)-1):
        time1[i+1]=time1[i]+h
        k1=h*y[i]           
        l1=h*(3*m*z[i]**2-z[i]-2*(q**2)*(z[i]**3))  
        k2=h*(y[i]+l1/2)      
        l2=h*(3*m*(z[i]+k1/2)**2-(z[i]+k1/2)-2*(q**2)*((z[i]+k1/2)**3))   
        k3=h*(y[i]+l2/2)      
        l3=h*(3*m*(z[i]+k2/2)**2-(z[i]+k2/2)-2*(q**2)*((z[i]+k2/2)**3))
        k4=h*(y[i]+l3)      
        l4=h*(3*m*(z[i]+k3/2)**2-(z[i]+k3/2)-2*(q**2)*((z[i]+k3/2)**3))
        z[i+1]=(z[i]+(1/6)*(k1+2*k2+2*k3+k4))
        y[i+1]=y[i]+(1/6)*(l1+2*l2+2*l3+l4)
        r[i+1]=1/z[i]

    x_graf=r*np.cos(time1)
    y_graf=r*np.sin(time1)
    for j in range(0,len(time1)-1):
        #if x_graf[j]<-l/2 or y_graf[j]<-l/2 or r[j]<1 or x_graf[j]>l/2 or y_graf[j]>l/2:
        if r[j]<rh:
            x_graf[j]=np.nan
            y_graf[j]=np.nan
    return x_graf,y_graf,time1
```
