## Gamma incompleta num√©rica r vs g00

### Funci√≥n para obtener theta para un agujero negro con un s√≥lo horizonte

```python
def theta_function(m,theta,tt):
    for k in range(20):    
        gamma_inc_inf=scipy.special.gamma(3/2)*scipy.special.gammainc(3/2, (tt**2)/(4*theta));
        g00=(1-4*m*gamma_inc_inf/(np.sqrt(np.pi)*tt))
        Dg00 = np.diff(g00)/np.diff(tt)
        for j in range(len(Dg00)):
            if Dg00[j]<0 and Dg00[j+1]>0 or Dg00[j]>0 and Dg00[j+1]<0:             
                rh_posicion=j
                break         
        rh=tt[rh_posicion]
        func = lambda tau : (1-4*m*(scipy.special.gamma(3/2)*scipy.special.gammainc(3/2, (rh**2)/(4*tau)))/(np.sqrt(np.pi)*rh))
        tau_initial_guess = 0.01
        tau_solution = fsolve(func, tau_initial_guess)
        if theta == tau_solution:
            break
        theta=tau_solution
    return (g00, Dg00, tau_solution, rh, rh_posicion)
```

```python
#LLAMADO A LA FUNCION 
m=.5
theta_in=1
tt=np.linspace(0.001, 7, 10000) 
g00=theta_function(m,theta_in,tt)[0]
Dg00=theta_function(m,theta_in,tt)[1]
theta=float(theta_function(m,theta_in,tt)[2])
rh=theta_function(m,theta_in,tt)[3]
rh_pos=theta_function(m,theta_in,tt)[4]
```

## Geod√©sicas

### Funciones para geodesicas de Schwarzchild y Schwarzchild no conmutativo (aproximaci√≥n)

```python
def geodesicas_aprox(d,l,rs,n,ti,tf,z0,y0,q,alpha): 
    h=(tf-ti)/n  
    time1=list(range(0,n+1));
    y=list(range(0,n+1)) 
    z=list(range(0,n+1))
    r=list(range(0,n+1))
    y[0]=y0
    z[0]=z0 
    r[0]=1/z0
    rs=1
    time1[0]=ti
    m=rs/2
    rh=rs/2+np.sqrt(rs**2-4*q**2)/2
    theta1=.06895
    
    for i in range(0,len(time1)-1):
        time1[i+1]=time1[i]+h
        k1=h*y[i]           
        l1=h*(3*m*z[i]**2-z[i]+alpha*((3*m*z[i])/np.sqrt(np.pi*theta1)*np.exp(-1/(4*theta1*z[i]))-m/(z[i]*np.sqrt(2*np.pi*theta1**3))*np.exp(-1/(4*theta1*z[i]))))
        k2=h*(y[i]+l1/2)      
        l2=h*(3*m*(z[i]+k1/2)**2-(z[i]+k1/2)+alpha*((3*m*(z[i]+k1/2))/np.sqrt(np.pi*theta1)*np.exp(-1/(4*theta1*(z[i]+k1/2)))-m/((z[i]+k1/2)*np.sqrt(2*np.pi*theta1**3))*np.exp(-1/(4*theta1*(z[i]+k1/2)))))
        k3=h*(y[i]+l2/2)      
        l3=h*(3*m*(z[i]+k2/2)**2-(z[i]+k2/2)+alpha*((3*m*(z[i]+k2/2))/np.sqrt(np.pi*theta1)*np.exp(-1/(4*theta1*(z[i]+k2/2)))-m/((z[i]+k2/2)*np.sqrt(2*np.pi*theta1**3))*np.exp(-1/(4*theta1*(z[i]+k2/2)))))
        k4=h*(y[i]+l3)      
        l4=h*(3*m*(z[i]+k3/2)**2-(z[i]+k3/2)+alpha*((3*m*(z[i]+k3/2))/np.sqrt(np.pi*theta1)*np.exp(-1/(4*theta1*(z[i]+k3/2)))-m/((z[i]+k3/2)*np.sqrt(2*np.pi*theta1**3))*np.exp(-1/(4*theta1*(z[i]+k3/2)))))
        z[i+1]=(z[i]+(1/6)*(k1+2*k2+2*k3+k4))
        y[i+1]=y[i]+(1/6)*(l1+2*l2+2*l3+l4)
        r[i+1]=1/z[i]

    x_graf=r*np.cos(time1)
    y_graf=r*np.sin(time1)
    for j in range(0,len(time1)-1):
        #if x_graf[j]<-l/2 or y_graf[j]<-l/2 or r[j]<1 or x_graf[j]>l/2 or y_graf[j]>l/2:
        if r[j]<rh:
            x_graf[j]=np.nan
            y_graf[j]=np.nan
    return x_graf,y_graf,time1
```

```python
def den_funcion(dat,malla_x,malla_y,L):
    x=dat[:,0]
    y=dat[:,1]
    z=dat[:,2]
    den=np.zeros([malla_x,malla_y])
    z_new=np.zeros([malla_x,malla_y])
    z_final=np.zeros([malla_x,malla_y])
    x=x*malla_x/L
    y=y*malla_x/L
    for i in range(len(x)):   
        if z[i]>1:
            z[i]=1
        a=int(int(x[i])+malla_x/2-1)
        b=-int(int(y[i])+malla_y/2-1)
        den[a,b]=den[a,b]+1
        z_new[a,b]=z_new[a,b]+z[i]
        z_final[b,a]=z_new[a,b]/den[a,b]
    return z_final
```

```python
r1 = np.linspace(0.5, 7, 10000)
v1=np.zeros(len(r1))
S=np.zeros(len(r1))
rs=2*m
for k in range(len(r1)):
    v1[k]=(1-rs/r1[k]-rs*np.exp(-((r1[k]**2)/(4*theta)))/(np.sqrt(np.pi*theta)))
    S[k]=(1-rs/r1[k])
```

```python
Or_SW=30
mymodel_SW = np.poly1d(np.polyfit(r1, S, Or_SW))
```

```python
tt_aux=np.linspace(0.5, 7, 10000)
g00_aux=np.zeros(10000)
for i in range (len(tt_aux)):
    aux=0
    for j in range (len(mymodel_SW)+1):
        aux=aux+((tt_aux[i]**j)*np.array(mymodel_SW[j]))
    g00_aux[i]=aux


    
plt.figure(figsize=(10,5))
#plt.ylim(-1.1, 1.1)
plt.xlabel("$r$",fontsize=15, color = 'black')
plt.ylabel("$g_{00}$",fontsize=15, color = 'black')
plt.plot(tt_aux,g00_aux, color='red',label="Te√≥rica Schwarzschild no conmutativo")
plt.plot(r1,S, color='blue',label="Te√≥rica Schwarzschild no conmutativo")
#plt.plot(tt,S, color='green',label="Te√≥rica Schwarzschild no conmutativo")
plt.grid()
plt.legend(loc='lower right')
```

```python
D_Linea_Tendencia_SW = np.diff(g00_aux)/np.diff(tt_aux)
D_SW= np.diff(S)/np.diff(r1)
```
